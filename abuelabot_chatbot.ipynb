```python
from unsloth import FastLanguageModel
import torch
import zipfile
import os
```

```python
## unzip model
zip_path = "/content/abuela-llm-v1.zip"
extract_dir = "/content/abuela-llm-v1"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("Extracted model files to:", extract_dir)
```

```python

base_model_name = "unsloth/Qwen2.5-7B"
lora_path = "/content/abuela-llm-v1"

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = base_model_name,
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
    adapter_name = lora_path,
)

FastLanguageModel.for_inference(model)
```

```python
model_path = "/content/abuela-llm-v1/abuela-llm-v1"

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=model_path,
    max_seq_length=2048,
    dtype=None,
    load_in_4bit=True,
)

FastLanguageModel.for_inference(model)
```

```python
import torch

device = model.device if hasattr(model, "device") else ("cuda" if torch.cuda.is_available() else "cpu")
if getattr(tokenizer, "pad_token_id", None) is None:
    tokenizer.pad_token_id = getattr(tokenizer, "eos_token_id", None)

SYSTEM = "Eres una abuelita puertorriqueña amable, sabia y simpática."

def ask_once(user_text: str):
    prompt = (
        f"<|system|>{SYSTEM}<|end|>"
        f"<|user|>{user_text}<|end|>"
        f"<|assistant|>"
    )
    enc = tokenizer(prompt, return_tensors="pt")
    enc = {k: v.to(device) for k, v in enc.items()}

    with torch.no_grad():
        out = model.generate(
            **enc,
            max_new_tokens=160,
            do_sample=True, temperature=0.7, top_p=0.95, top_k=50,
            repetition_penalty=1.1,
            pad_token_id=tokenizer.pad_token_id,
            use_cache=True,
        )

    prompt_len = enc["input_ids"].shape[-1]
    gen = out[0][prompt_len:]
    return tokenizer.decode(gen, skip_special_tokens=True).strip()

print(ask_once("Abuela, me duele la garganta. ¿Qué me tomo?"))
```

```python
## testing

print(ask_once("¿Cómo hago arroz con gandules para 4 personas?"))
print(ask_once("¿I failed an exam, how do I tell mami?"))
print(ask_once("Hi Abuela, how are you doing today?"))
print(ask_once("Me siento mal, me duele la garganta."))
```
